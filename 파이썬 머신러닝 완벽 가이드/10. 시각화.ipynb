# 1. 시각화를 시작하며 - 맷플롯립과 시본 개요

- 맷플롯립(Matplotlib)
    - 파이썬 시각화에 큰 공헌을 한 시각화 라이브러리
    - 직관적이지 못한 개발 API
    - 부가적인 속성까지 코딩 필요
    - 현대적인 감각이 떨어지는 시각화 플롯
- 시본(Seaborn)
    - 쉬운 구현, 수려한 시각화, 편리한 연동
    - 상대적으로 적은 양의 코딩으로 시각화 가능
    - 맷플롯립 이해 필요

# 2. 맷플롯립(Matplotlib)

- `pyplot` 모듈
    - `plot()` : 선 그래프
    - `title()` : 차트 타이틀 설정
    - `show()` : 선 그래프 화면 표시
    - `figure()` : Figure 객체 반환
    - `axes()` : 현재 사용하는 Axes 객체 반환
    - `subplots()` : 여러 개의 subplot 생성
    - `xlabel()` : X축 명칭
    - `ylabel()` : Y축 명칭
    - `xticks()` : X축 눈금값 회전
    - `yticks()` : Y축 눈금값 회전
    - `legend()` : 범례 표시
- Figure 객체
    - 그림을 그리기 위한 캔버스 역할
    - 전체 그림판 크기 조절
- Axes
    - 실제 그림을 가지는 메서드
    - 대부분의 시각화 수행
    - X축, Y축, 타이틀, 범례 등 속성 설정

# 3. 시본(Seaborn)

- 히스토그램
    - 연속형 값에 대한 도수 분포 표현
    - `histplot()` : Axes 레벨 함수
    - `displot()` : Figure 레벨 함수
- 바 플롯
    - 특정 칼럼의 이산값에 따른 다른 칼럼의 연속형 값을 막대 그래프 형태로 시각화
    - `barplot()`
- 박스 플롯
    - 연속형 값의 사분위 IQR과 최대, 최소, 이상치 값 시각화
    - `boxplot()`
- 바이올린 플롯
    - 히스토그램 연속 확률 분포 곡선과 박스 폴롯을 바이올린 형태로 함께 시각화
    - `violinplot()`
- 스캐터 플롯
    - 산점도, 2개의 연속형 값들을 X, Y 좌표상의 점으로 시각화
    - `scatterplot()`
- 상관 히트맵
    - 다수의 연속형 칼럼들에 대해 상관 관계 시각화
    - `heatmap()`
- 카운트 플롯
    - 이산형 값 건수 막대그래프 형태로 시각화
    - `countplot()`

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')
ax[0].legend()
ax[1].legend()
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')
plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()
plt.show()
```
