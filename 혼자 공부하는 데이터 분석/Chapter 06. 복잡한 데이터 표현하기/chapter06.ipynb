```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
top30_pubs = ns_book7[''].value_counts()[:30]
top30_pubs_idx = ns_book7[''].isin(top30_pubs.index)
```

```python
ns_book9 = ns_book7[top30_pubs_idx][['', '', '']]
```

```python
ns_book9 = ns_book9.groupby(by=['', '']).sum()
```

```python
ns_book9 = ns_book9.reset_index()
ns_book9[ns_book9[''] == ''].head()
```

```python
line1 = ns_book9[ns_book9[''] == '']
line2 = ns_book9[ns_book9[''] == '']
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(line1[''], line1[''])
ax.plot(line2[''], line2[''])
ax.set_title(' ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(line1[''], line1[''], label='')
ax.plot(line2[''], line2[''], label='')
ax.set_title(' ')
ax.legend()
fig.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for pub in top30_pubs.index[:5]:
    line = ns_book9[ns_book9[''] == pub]
    ax.plot(line[''], line[''], label=pub)
ax.set_title(' ')
ax.legend()
ax.set_xlim(1985, 2025)
fig.show()
```

```python
ns_book10 = ns_book9.pivot_table(index='', columns='')
ns_book10.head()
```

```python
ns_book10.columns[:10]
```

```python
top10_pubs = top30_pubs.index[:10]
year_cols = ns_book10.columns.get_level_values(1)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.stackplot(year_cols, ns_book10.loc[top10_pubs].fillna(0), labels=top10_pubs)
ax.set_title(' ')
ax.legend(loc='upper left')
ax.set_xlim(1985, 2025)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(line1[''], line1[''], label='')
ax.bar(line2[''], line2[''], label='')
ax.set_title(' ')
ax.legend()
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(line1[''] - 0.2, line1[''], width=0.4, label='')
ax.bar(line2[''] + 0.2, line2[''], width=0.4, label='')
ax.set_title(' ')
ax.legend()
fig.show()
```

```python
height1 = [5, 4, 7, 9, 8]
height2 = [3, 2, 4, 1, 2]
plt.bar(range(5), height1, width=0.5)
plt.bar(range(5), height2, bottom=height1, width=0.5)
plt.show()
```

```python
height3 = [a + b for a, b in zip(height1, height2)]
plt.bar(range(5), height3, width=0.5)
plt.bar(range(5), height1, width=0.5)
plt.show()
```

```python
ns_book10.loc[top10_pubs[:5], ('', 2013):('', 2020)]
```

```python
ns_book10.loc[top10_pubs[:5], ('', 2013):('', 2020)].cumsum()
```

```python
ns_book12 = ns_book10.loc[top10_pubs].cumsum()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for i in reversed(range(len(ns_book12))):
    bar = ns_book12.iloc[i]
    label = ns_book12.index[i]
    ax.bar(year_cols, bar, label=label)
ax.set_title(' ')
ax.legend(loc='upper left')
ax.set_xlim(1985, 2025)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(data, labels=labels, startangle=90, autopct='%.1f%%', explode=[0.1]+[0]*9)
ax.set_title('  ')
fig.show()
```
